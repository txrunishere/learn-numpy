```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array(
  [
    [1, 15000, 18000, 22000, 25000],
    [2, 12000, 14000, 16000, 19000],
    [3, 20000, 23000, 26000, 30000],
    [4, 18000, 21000, 24000, 27000],
    [5, 16000, 18500, 20500, 23000]
  ]
)
```

```python
# print("Data of first 3 rows:\n", sales_data[:3])
# print(sales_data[:,:])
print(sales_data[:, 1]) # column at 1 index
print(sales_data[:, 1:]) # all column after 0 index
```

```python
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("Total per year: ", yearly_total)
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimun sales of a restaurant: ", min_sales)
max_sales = np.max(sales_data[:, 1:], axis=1)
print("Maximun sales of a restaurant: ", max_sales)

# Maximun sales per year
max_sales_year = np.max(sales_data[:, 1:], axis=0)
print(max_sales_year)

# Average sales per restaurant
avg_sales_restaurant = np.average(sales_data[:, 1:], axis=1)
# avg_sales_restaurant = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales_restaurant)
```

```python
# cumulative sum
cumulative_sum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumulative_sum)

avg_cumsum = np.mean(cumulative_sum, axis=0)
print(avg_cumsum)
plt.figure(figsize=(12, 6))
plt.plot(avg_cumsum)
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
